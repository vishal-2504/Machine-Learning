## **Q.1)**

```python
# Provided code for generating data
x = np.arange(0, 20.1, 0.1)
np.random.seed(0)
y = 1*x**5 + 3*x**4 - 100*x**3 + 8*x**2 -300*x - 1e5 + np.random.randn(len(x))*1e5
```

```python
# Function for linear-regression
def lin_regression(X,Y):
  lin_reg = LinearRegression()
  lin_reg.fit(X,Y)
  return lin_reg.intercept_,lin_reg.coef_
```

```python
# Transforming to linear regression, p=4 (polynomial of degree 4)

x1 = x
x2 = x**2
x3 = x**3
x4 = x**4
x_mat = np.ones((len(x),4))
x_mat[:,0] = x1
x_mat[:,1] = x2
x_mat[:,2] = x3
x_mat[:,3] = x4
intercept, coeff = lin_regression(x_mat,y)
print('Intercept Obtained',intercept)
print('Coefficient obtained',coeff)
y_predicted = coeff[0]*x +  coeff[1]*x**2 + coeff[2]*x**3 + coeff[3]*x**4 + intercept
rms = mean_squared_error(y, y_predicted, squared=False) 
print('RMS error', rms)
```

```python
# For p=5 (polynomial of degree 5)

x5 = x**5
x_mat = np.ones((len(x),5))
x_mat[:,0] = x1
x_mat[:,1] = x2
x_mat[:,2] = x3
x_mat[:,3] = x4
x_mat[:,4] = x5
intercept, coeff = lin_regression(x_mat,y)
print('Intercept Obtained',intercept)
print('Coefficient obtained',coeff)
y_predicted = coeff[0]*x +  coeff[1]*x**2 + coeff[2]*x**3 + coeff[3]*x**4 + coeff[4]*x**5 + intercept
rms = mean_squared_error(y, y_predicted, squared=False) 
print('RMS error', rms)
```

```python
# L1, finding minima using gradient descent method
# for alpha 0.01
alpha = 0.01
theta = 10.0     #starting with initial value 10
epsilon = 1e-6   
itr=0
while 1:
  t1 = theta - alpha*2*theta
  itr+=1
  if abs(theta-t1)<epsilon:      # Convergence condition
    theta=t1
    break
  theta=t1
print('Number of iterations:',itr)
print('Minimum value is:',theta*theta)
print('at theta:',theta)
```

# **2-e)**

```python
#function to compute partial differentiation w.r.t theta0
def diff_theta0(x,y,theta0,theta1):
  diff = 0
  for i in range(len(x)):
    diff=diff+theta0+(theta1*x[i])-y[i]
  return 2*diff
```

```python
#function to compute partial differentiation w.r.t theta1
def diff_theta1(x,y,theta0,theta1):
  diff=0
  for i in range(len(x)):
    diff=diff+(theta0+(theta1*x[i])-y[i])*x[i]
  return 2*diff
```

```python
#using data-file provided

alpha = 0.01
data = pd.read_excel('data_1.xlsx')   
x=np.array(data['x'])
x=(x-np.mean(x))/np.std(x)                  #doing standardization so that the values don't shoot-up
y=np.array(data['y'])
theta = np.array([1.0,1.0])
epsilon = 1e-6
itr=0
while 1:
  t0 = theta[0]-alpha*diff_theta0(x,y,theta[0],theta[1])
  t1 = theta[1]-alpha*diff_theta1(x,y,theta[0],theta[1])
  itr+=1
  if abs(theta[0]-t0)<epsilon and abs(theta[1]-t1)<epsilon:       #condition for convergence
    theta[0] = t0
    theta[1] = t1
    break
  theta[0] = t0
  theta[1] = t1

cost = 0
for i in range(len(x)):
  cost+=(theta[0]+theta[1]*x[i]-y[i])**2
print('Number of iterations:',itr)
print('Minimum value is:',cost)
print('at theta vector:', theta)
```

```python
def diff_theta0_sgd(x,y,theta0,theta1):
  diff = 0
  diff=diff+theta0+(theta1*x)-y
  return 2*diff
```

```python
#function to compute partial differentiation w.r.t theta1
def diff_theta1_sgd(x,y,theta0,theta1):
  diff=0
  diff=diff+(theta0+(theta1*x)-y)*x
  return 2*diff
```

```python
#stochastic gradient descent
#calculating the no of iterations for convergence for different alpha values

alpha_arr = np.array([0.01,0.03,0.08,0.1,0.15])
data = pd.read_excel('data_1.xlsx')   
x=np.array(data['x'])
x=(x-np.mean(x))/np.std(x)   #doing standardization so that the values don't shoot-up
y=np.array(data['y'])
epsilon = 1e-6
no_itr = np.zeros(len(alpha_arr))
minima = np.zeros(len(alpha_arr))
i=0
for alpha in alpha_arr:
  theta = np.array([1.0,1.0])
  itr=0
  while 1:     #the iteration do not lead to convergence that's why putting another condition in-order to terminate the loop
    r = np.random.randint(0,len(y))   #taking random integer to take random sample to evaluate gradient
    t0 = theta[0]-alpha*diff_theta0_sgd(x[r],y[r],theta[0],theta[1])
    t1 = theta[1]-alpha*diff_theta1_sgd(x[r],y[r],theta[0],theta[1])
    itr+=1
    if abs(theta[0]-t0)<epsilon and abs(theta[1]-t1)<epsilon:       #condition for convergence
      theta[0] = t0
      theta[1] = t1
      break
    theta[0] = t0
    theta[1] = t1
  no_itr[i] = itr
  cost = 0
  for it in range(len(x)):
    cost+=(theta[0]+theta[1]*x[it]-y[it])**2
  minima[i] =  cost
  i+=1

#printing the results in form of table
data = {'Learning-Rate':  alpha_arr, 'No of Iterations': no_itr,'Minima Obtained':minima}
df = pd.DataFrame (data, columns = ['Learning-Rate','No of Iterations','Minima Obtained'])
df
```

```python
#using steepest descent
#a)
theta = 10.0     #starting with initial value 10
epsilon = 1e-6   
itr=0
grad = 1.0
while grad>=epsilon:
  alpha = 1/2                       #alpha is determined for each iteration, turns out to be 0.5 in this case           
  grad =  2*theta        
  theta = theta - alpha*grad
  itr+=1

print('Number of iterations:',itr)
print('Minimum value is:',theta*theta)
print('at theta:',theta)
```

```python
#b)

theta = np.array([10.0,10.0])
epsilon = 1e-6   
itr=0
grad = np.ones((2,1))
while np.sqrt(np.matmul(grad.T,grad)):              #converging condition according to algorithm
  alpha=1/2
  grad[0][0] = 2*theta[0]
  grad[1][0] = 2*theta[1]
  theta[0] = theta[0]-alpha*grad[0][0]
  theta[1] = theta[1]-alpha*grad[1][0]
  itr+=1
print('Number of iterations:',itr)
print('Minimum value is:',theta[0]**2+theta[1]**2)
print('at theta vector:', theta)
```

```python
#function to calculate cost
def calculate_cost(theta, x, y):
  cost = 0.0
  for i in range(len(x)):
    cost+=(theta[0]+(theta[1]*x[i])-y[i])**2
  return cost
```

```python
#function to calculate alpha for steepest descent method for given function
#equation is derived manually
def calculate_alpha(theta0, theta1, x, y, gradient):
  g0 = gradient[0][0]
  g1 = gradient[1][0]
  temp = 0.0
  temp2 = 0.0
  for i in range(len(x)):
    temp += g0*(theta0+theta1*x[i]-y[i]) + g1*(theta0*x[i]+theta1*x[i]**2-x[i]*y[i])
    temp2 += (g0**2)+(2*g0*g1*x[i])+(g1**2)*(x[i]**2)
  return temp/temp2
```

```python
#steepest descent

data = pd.read_excel('data_1.xlsx')   
x=np.array(data['x'])
x=(x-np.mean(x))/np.std(x)                  #doing standardization so that the values don't shoot-up
y=np.array(data['y'])
theta = np.array([1.0,1.0])
epsilon = 1e-6
itr=0
grad = np.ones((2,1))
while np.sqrt(np.matmul(grad.T,grad))>=epsilon:
  grad[0][0] = diff_theta0(x,y,theta[0],theta[1])
  grad[1][0] = diff_theta1(x,y,theta[0],theta[1])
  alpha = calculate_alpha(theta[0], theta[1], x, y, grad)
  theta[0] = theta[0]-(alpha*grad[0][0]) 
  theta[1] = theta[1]-(alpha*grad[1][0])
  itr+=1

cost = calculate_cost(theta, x, y)
print('Number of iterations:',itr)
print('Minimum value is:',cost)
print('at theta vector:', theta)
```
